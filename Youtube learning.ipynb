## Creating Long strategy only, no shorting

```python

data["position2"]=np.where(data["sma_50"]>data["sma_100"],1,-0)
```

```python


"""
What it does:
For each day, if you were “in” (according to your signal) yesterday, you get today’s return; if you were “out,” you get 0.
Why:
This simulates your strategy: only earn returns when your system says to be in the stock.
    
    
"""
data["strategy2"] = data["returnsb&h"] * data["position2"].shift(1)
```

```python
data.dropna(inplace=True)   
```

```python
"""
What it does:
Adds up all daily log returns for both:
Buy & Hold (always in the stock)
Your Strategy (only in when your rule says so)
Why:
This gives the total log return for each approach over the entire period.

Output:
Output is log return not percentage
"""

data[["returnsb&h","strategy2"]].sum()
```

```python
"""
What it does:
Converts the total log return into a growth factor (e.g., if result is 4, your money grew 4×).
Why:
Log returns are additive, but not intuitive. Taking the exponential (np.exp) turns this into something you can interpret as “my money multiplied by X.”

Log return is calculated as log(today’s price / yesterday’s price).
You can add up log returns over many periods to get a “total log return.”
But the sum of log returns is not a plain percentage—it’s a number you have to “convert” to a percentage using the exponential function.


"""

# directly comparing strategy1 and 2
data[["strategy","strategy2"]].sum().apply(np.exp) # what 1$ will be worth if invested at the start of the period
```

## Creating Function


```python
def test_strategy(stock,start,end,SMA):
    df = yf.download(stock, start=start, end=end)
    data = df["Close"][f'{stock}'].to_frame()
    #use f-string becoz current download structure is different    
    data['returns'] = np.log(data[f'{stock}'].div(data[f'{stock}'].shift(1)))
    data['SMA_S'] = data[f"{stock}"].rolling(int(SMA[0])).mean()
    data['SMA_L'] = data[f"{stock}"].rolling(int(SMA[1])).mean()
    data.dropna(inplace=True)
    
    data["position"]=np.where(data["SMA_S"]>data["SMA_L"],1,0)
    data["ret_strategy"]=data["returns"]*data.position.shift(1)
    data.dropna(inplace=True)
    ret = np.exp(data["ret_strategy"].sum())
    std = data["ret_strategy"].std()*np.sqrt(252)
    
    return ret, std


#assumption: Just long position and using SMA 50,200 only
```

## Creating 1st class

```python
class SMABacktester():
    def __init__(self, stock ,SMA_S,SMA_L,start, end):
        self.stock = stock
        self.SMA_S = SMA_S
        self.SMA_L = SMA_L
        self.start = start
        self.end = end
        self.results = None #placeholder for now
        self.get_data()
    
    def get_data(self):
        df = yf.download(self.stock, start=self.start, end=self.end)
        print(f"\n{df.info()}\n")
        data = df["Close"][f'{self.stock}'].to_frame() 
        data['returns'] = np.log(data[f'{self.stock}'].div(data[f'{self.stock}'].shift(1)))
        data['SMA_S'] = data[f"{self.stock}"].rolling(int(self.SMA_S)).mean()
        data['SMA_L'] = data[f"{self.stock}"].rolling(int(self.SMA_L)).mean()
        data.dropna(inplace=True)
        self.data2 = data
        return self.data2
        
    def test_results(self):
        data = self.data2.copy()
        data["position"]=np.where(data["SMA_S"]>data["SMA_L"],1,0)
        
        #ret_startegy is the return for every interval (D, W, Month etc) if D then daily reuturn
        data["ret_strategy"]=data["returns"] * data.position.shift(1)
        data.dropna(inplace=True)
        
        # this column shows buy and hold
        data["returnsbh"]=data["returns"].cumsum().apply(np.exp)
        
        #Strategy Column shows the cul return overtime of this strategy
        data["strategy"]=data["ret_strategy"].cumsum().apply(np.exp)
        perf_strategy=data["strategy"].iloc[-1]
        perf_simplybh = data["returnsbh"].iloc[-1]
        outperf=perf_strategy - perf_simplybh
        self.results = data
        
        # ret = np.exp(data["ret_strategy"].sum())
        # std = data["ret_strategy"].std()*np.sqrt(252)
        
        return f'strategy: {perf_strategy.round(6)}', f'bh: {perf_simplybh.round(6)}', round(outperf,6)
        
    def Plot_result(self):
        if self.results is None:
            print("No results to plot yet. Run test_results() first.")
        else:
            #plotting the cumulative return graph of both strategy
            ax = self.results[["returnsbh", "strategy"]].plot(
                figsize=(12, 8), 
                fontsize=15, 
                color=["blue", "red"]   # returnsbh is blue, strategybh is red
            )
            ax.set_title(f"{self.stock} | SMA{self.SMA_S} & SMA{self.SMA_L}", fontsize=20)
            plt.show()
```
